```python
import csv
import numpy as np
import matplotlib.pyplot as plt

def read_shuffle_csv(file_name):
    f = open(file_name,'r',encoding='UTF8')
    rdr = csv.reader(f)
    Line = []
    for line in rdr:
      
      file_name = line[0].split('/')[-1]
      print(file_name)
      if line[0].split('_')[-1] != '1.png':
        Line.append([ file_name ,*line[1:]])
    f.close()
    return Line

def find_std(lines):
  std_dict = {}
  stds= []
  for line in lines:
    # print(line)
    delta_rotations = np.array(line[1:10], dtype=np.float64)
    print(delta_rotations)
    std = np.std(delta_rotations)
    print(std)
    std_dict[line[0]] = std
    stds.append(std)
  return std_dict, stds

def draw_std_histogram(stds):
  plt.hist(stds, bins=100, density=True, alpha=0.7, histtype='step')
  plt.show()

def draw_specific_img_histogram(lines):
  for i, line in enumerate(lines):
      x = np.arange(9)   
      delta_rotations = np.array(line[1:10], dtype=np.float64)
      plt.bar(x,delta_rotations, width=0.5)
      plt.xticks(x,[a for a in range(9)] )
      plt.savefig(f'./imageRotHist/{line[0]}.png')
      plt.close()
  




Line = read_shuffle_csv('histogram.csv')
std_dict, stds = find_std(Line)
print(std_dict)
```

```python
draw_specific_img_histogram(Line)
```
